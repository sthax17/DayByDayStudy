```python
import pandas as pd
class WRMSE:
    def __init__(self):
        self.df = pd.DataFrame()
        self.sum_weight = 0
        self.WRMSE = 0
    def generate_df(self):
        batter = ['A', 'B', 'C', 'D']
        ab = [200, 50, 0, 0]
        ops = [0.9, 0.4, 0.0, 0.0]
        pred = [0.95, 0.35, 0.5, 0.3]
        dic = {'Batter':batter, 'AB':ab, 'OPS':ops, 'Pred':pred}
        self.df = pd.DataFrame(dic)
    def sigma_w(self):
        for i in self.df:
            self.sum_weight += self.df.loc[i, 'AB']
    def get_WRMSE(self):
        self.generate_df()
        self.sigma_w()
        for i in self.df:
            diff = self.df.loc[i, 'OPS'] - self.df.loc[i, 'Pred']
            diff_powered = diff ** 2
            temp = (diff_powered*self.df.loc[i, 'AB'])/self.sum_weight
            self.WRMSE += temp ** (1/2)
            print(self.WRMSE)
wrmse = WRMSE()
wrmse.get_WRMSE()
```

```python
import numpy as np

AB = np.array([200,50,0,0])
OPS = np.array([0.9,0.4,0.0,0.0])
Pred = np.array([0.95,0.35,0.5,0.3])

class Metrics():
    def __init__(self, ops, pred, ab):
        self.ops = ops 
        self.pred = pred
        self.ab = ab

def out_wrmse(self):
    wrmse_value = \
    np.sqrt(((self.ops - self.pred) ** 2) * self.ab / np.sum(self.ab))
    wrmse_total = np.round(np.sum(wrmse_value),6)
    return wrmse_total

Metrics.wrmse= out_wrmse

C_wrmse = Metrics(OPS, Pred, AB)

print("Wrmse_value : ", C_wrmse.wrmse())
```

```python
faults_norm = (faults - faults.min()) / (faults.max() - faults.min()) 
```

```python
display(faults_norm.describe())
```

```python
print(faults_norm[:5])
faults_shuffle = faults_norm.sample(frac=1)
print("==============================")
print(faults_shuffle[:5])
```

```python
faults_np = faults_shuffle.to_numpy()
print(type(faults_np))
```

```python
train_ratio = 0.8
input_cnt   = 4
output_cnt  = 3

train_idx = int(len(faults_np) * train_ratio)
print("train_idx : ", train_idx)
```

```python
train_X, train_Y = faults_np[:train_idx, :-output_cnt], faults_np[:train_idx, -output_cnt : ]
test_X, test_Y   = faults_np[train_idx:, :-output_cnt], faults_np[train_idx:, -output_cnt : ]
```

```python
print("train_X.shape : ", train_X.shape)
print("train_Y.shape : ", train_Y.shape)

print("test_X.shape : ", test_X.shape)
print("test_Y.shape : ", test_Y.shape)
```

```python
simple_layer_model = tf.keras.Sequential([
                                          tf.keras.layers.Dense(units = 3, 
                                                                activation  = 'softmax',
                                                                input_shape = (4,))
])
```

```python
simple_layer_model.summary()
```

```python
simple_layer_model.compile(optimizer = tf.keras.optimizers.Adam(learning_rate=0.001),
                           loss = 'categorical_crossentropy')
history_1 = simple_layer_model.fit(train_X, train_Y,
                                   epochs = 25, batch_size = 2)
```

```python
simple_layer_model.evaluate(test_X, test_Y)
```

```python
plt.figure(figsize=(12,4))
plt.plot(history_1.history['loss'], 'b--o', label = 'loss')
plt.xlabel("Epochs")
plt.ylabel("Loss")
plt.legend()
plt.grid()
plt.show()
```

```python
pred_Y = simple_layer_model.predict(test_X)
print("pred_Y : \n", pred_Y)
```

```python
pred_Y = np.argmax(pred_Y, axis = 1)
print("pred_Y : ", pred_Y)
```

```python
label_Y = np.argmax(test_Y,axis = 1)
print("label_Y \n", label_Y)
```

```python
acc = np.mean(np.equal(pred_Y, label_Y))
print("Acc : {} % ".format(np.round(acc * 100,2)))
```
