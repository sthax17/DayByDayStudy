```python
%cd /
from glob import glob

img_test_list = glob('/content/dataset/test/images/*.jpg')
img_train_list = glob('/content/dataset/train/images/*.jpg')
img_valid_list = glob('/content/dataset/valid/images/*.jpg')
print(len(img_test_list))
print(len(img_train_list))
print(len(img_valid_list))
```

```python
with open('/content/dataset/test.txt', 'w') as f:
    f.write('\n'.join(img_test_list) + '\n')    
with open('/content/dataset/train.txt', 'w') as f:
    f.write('\n'.join(img_train_list)  + '\n')
with open('/content/dataset/val.txt', 'w') as f:
    f.write('\n'.join(img_valid_list) + '\n')    
```

```python
import yaml


with open('/content/dataset/data.yaml', 'r') as f:
    data = yaml.load(f)

print(data)

data['train'] = '/content/dataset/train.txt'
data['val'] = '/content/dataset/val.txt'

with open('/content/dataset/data.yaml', 'r') as f:
    yaml.dump(data, f)

print(data)
```
